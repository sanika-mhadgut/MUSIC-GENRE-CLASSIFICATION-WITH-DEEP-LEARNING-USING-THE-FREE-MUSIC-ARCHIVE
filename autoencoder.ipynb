# autoencoder

```python
import numpy as np
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import LabelEncoder
from torch.utils.data import DataLoader, TensorDataset
import pandas as pd
import torch.nn as nn

class MusicDataset(Dataset):
    def __init__(self, features, labels):
        label_encoder = LabelEncoder()
        labels_encoded = label_encoder.fit_transform(labels)

        self.features = torch.tensor(features.to_numpy(), dtype=torch.float32) if isinstance(features, pd.DataFrame) else torch.tensor(features, dtype=torch.float32)
        self.labels = torch.tensor(labels_encoded, dtype=torch.long)

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        return self.features[idx], self.labels[idx]

train_dataset = MusicDataset(X_train, y_train)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
```

```python
autoencoder_losses = []

class Autoencoder(nn.Module):
    def __init__(self, input_size):
        super(Autoencoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_size, 256),
            nn.LeakyReLU(0.1),
            nn.Linear(256, 128),
            nn.LeakyReLU(0.1),
            nn.Linear(128, 64),
            nn.LeakyReLU(0.1),
            nn.Linear(64, 32)
       )

        self.decoder = nn.Sequential(
            nn.Linear(32, 64),
            nn.LeakyReLU(0.1),
            nn.Linear(64, 128),
            nn.LeakyReLU(0.1),
            nn.Linear(128, 256),
            nn.LeakyReLU(0.1),
            nn.Linear(256, input_size),
            nn.Sigmoid()
        )

    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded

autoencoder = Autoencoder(input_size=X_train.shape[1])

optimizer = torch.optim.Adam(autoencoder.parameters(), lr=0.003)
criterion = nn.MSELoss()
num_epochs = 50

for epoch in range(num_epochs):
    total_loss = 0
    for inputs, _ in train_loader:
        inputs = inputs.to(device)
        optimizer.zero_grad()
        outputs = autoencoder(inputs)
        loss = criterion(outputs, inputs)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    avg_loss = total_loss / len(train_loader)
    autoencoder_losses.append(avg_loss)
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}')
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
label_encoder.fit(np.concatenate([y_train, y_test]))

y_test_encoded = label_encoder.transform(y_test)

y_test_tensor = torch.tensor(y_test_encoded, dtype=torch.long)

X_test_tensor = torch.tensor(X_test.to_numpy(), dtype=torch.float32) if isinstance(X_test, pd.DataFrame) else torch.tensor(X_test, dtype=torch.float32)
X_test_tensor = X_test_tensor.to(device)

autoencoder.eval()
with torch.no_grad():
    X_test_encoded = autoencoder.encoder(X_test_tensor)
```

```python
encoded_test_dataset = TensorDataset(X_test_encoded, y_test_tensor)
encoded_test_loader = DataLoader(encoded_test_dataset, batch_size=32, shuffle=False)
```

```python
class EncodedDataset(Dataset):
    def __init__(self, dataset, encoder):
        super(EncodedDataset, self).__init__()
        self.features = []
        self.labels = []
        self.encoder = encoder
        with torch.no_grad():
            for i in range(len(dataset)):
                feature, label = dataset[i]
                feature = feature.unsqueeze(0)
                encoded_feature = encoder(feature)
                self.features.append(encoded_feature.squeeze(0))
                self.labels.append(label)

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        return self.features[idx], self.labels[idx]

encoded_train_dataset = EncodedDataset(train_dataset, autoencoder.encoder)
encoded_train_loader = DataLoader(encoded_train_dataset, batch_size=32, shuffle=True)
```

```python
classifier_losses = []
classifier_accuracies = []

class Classifier(nn.Module):
    def __init__(self, num_features, num_classes):
        super(Classifier, self).__init__()
        self.layers = nn.Sequential(
            nn.Linear(num_features,50),
            nn.BatchNorm1d(50),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(50, num_classes)
        )
    def forward(self, x):
        return self.layers(x)

num_features = 32
num_classes = len(np.unique(y_train))
classifier = Classifier(num_features, num_classes).to(device)

classifier_optimizer = torch.optim.Adam(classifier.parameters(), lr=0.005, weight_decay=0.01)
classifier_criterion = nn.CrossEntropyLoss()

for epoch in range(30):
    epoch_loss = 0
    correct = 0
    total = 0
    for features, labels in encoded_train_loader:
        features, labels = features.to(device), labels.to(device)
        classifier_optimizer.zero_grad()
        outputs = classifier(features)
        loss = classifier_criterion(outputs, labels)
        loss.backward()
        classifier_optimizer.step()
        epoch_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
    epoch_accuracy = 100 * correct / total
    avg_loss = epoch_loss / len(encoded_train_loader)
    classifier_losses.append(avg_loss)
    classifier_accuracies.append(epoch_accuracy)
    print(f'Epoch {epoch+1}, Loss: {avg_loss:.4f}, Accuracy: {epoch_accuracy:.2f}%')
```

```python
def evaluate_model(model, data_loader):
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for data in data_loader:
            inputs, labels = data
            print("Processing a new batch")
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            print(f"Batch processed, Total: {total}, Correct: {correct}")

    accuracy = 100 * correct / total
    print(f'Accuracy: {accuracy:.2f}%')
```

```python
classifier.to(device)
evaluate_model(classifier, encoded_test_loader)
```
